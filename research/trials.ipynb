# Importing the data from drive

```python
import gdown
```

```python
url = "https://drive.google.com/file/d/14qbcWr-mgHMvbYgJmIN9HUBJPGeYaain/view?usp=sharing"
url.split("/")
```

```python
file_id = url.split("/")[-2]
```

```python
prefix = "https://drive.google.com/uc?/export=download&id="
gdown.download(prefix+file_id, "../data.zip")

```
